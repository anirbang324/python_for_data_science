seaborn is generally renamed as sns

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

there are several inbuilt datasets in seaborn library. 'iris' is one of them. To see how the dataset looks like we should use - ***variable.head()***

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.displot(iris['petal_length'],bins = 10)
```

```python
sns.kdeplot(iris['petal_length'])
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='reg')
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='petal_length',data=iris)
plt.show()
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.shape
```

```python
tips.head()
```

```python
sns.countplot(x=tips['sex'])
# sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.countplot(x=tips['day'].unique())
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.mean)
```

```python
sns.barplot(x='time',y='total_bill',data=tips,hue='day')
plt.show()
```
