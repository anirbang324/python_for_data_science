creating dummy data for normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_snd = np.random.randn(500) # standard deviation
X_snd
```

```python
maths_mu = 70
maths_sig = 9
maths_marks = np.round(X_snd*maths_sig+maths_mu)
```

```python
plt.hist(maths_marks, bins = 10)  #  Bins define the number of equal-width bins in the range.
plt.show()
```

```python
phy_mu = 50
phy_sig = 10
phy_marks = np.round(X_snd*phy_sig+phy_mu)
plt.hist(phy_marks)
plt.show()
```
