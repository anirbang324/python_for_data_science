# Scatter Plot
1. plt.scatter()
2. scatters data points across the x and y axis.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
```

```python
plt.scatter(x,y1, color='green',marker = 'o', label = "iPhone") 
plt.scatter(x,y2, color = 'red', label = "Android")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.title("Price of mobile devices over time")
plt.show()  
```

```python
dt = np.random.randn(100,2)
```

```python
plt.scatter(x = dt[:,0], y = dt[:,1],color = "red")
plt.show()
```
