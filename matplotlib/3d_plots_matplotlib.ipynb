```python
# importing numpy package from
# python library
import numpy as np

# importing matplotlib.pyplot package from
# python
import matplotlib.pyplot as plt

# Creating an empty figure
# or plot
fig = plt.figure()

# Defining the axes as a
# 3D axes so that we can plot 3D
# data into it.
ax = plt.axes(projection="3d")

# Showing the above plot
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(a,b,a+b)
plt.show()
```

```python
a = np.arange(-1,1,0.005)
b = a
a,b = np.meshgrid(a,b)
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(a,b,a**2 + b**2)
plt.show()
```
