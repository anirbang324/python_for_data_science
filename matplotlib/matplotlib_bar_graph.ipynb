```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(4)
car1 = [56,78,12,53]
car2 = [67,45,32,89]
```

Drawing two bar graphs 

```python
plt.bar(x,car1,width = 0.5,tick_label = [2012,2014,2016,2022],label = "car1")
plt.bar(x+0.5,car2,width = 0.3,label = "car2") #shifting the bar to the right so they won't overlap
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in million dollars")
plt.title("Cars price per year")
plt.show()
```
