```python
import matplotlib.pyplot as plt
import numpy as np
```

# Line Plots

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()  # to show the graph
```

```python
plt.style.available  # shows the available style
```

```python
plt.style.use('seaborn-v0_8-poster') # changing the background
```

# applying different marker styles and changing the color of the lines.
# specifying the title of the graph and adding a legend

# x is not a mandatory argument in matplotlib, only y is.

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
plt.plot(x,y1, color='green',marker = 'o', label = "iPhone") 
plt.plot(x,y2, color = 'red',marker="*", markersize = 12, linestyle = "dashed", label = "Android")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.title("Price of mobile devices over time")
plt.show()  
```
