Face Detection using OpenCV

# approach - we need to create a template, which is mostly the structure of a face
# using the template we can try to match the faces
# if the face is matching and then we can return that specific part from the window
# The technique is known as 'Sliding Window Technique'

# in openCV, we can use 'haarcascade.xml'. It is a template that contains so many face images.
# So, We need to download the file and put this in our folder.
# link : https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml

```python
import cv2
import matplotlib.pyplot as plt
```

```python
p = plt.imread("./grpimage.jpg")
plt.imshow(p)
```

```python
# cv2.CascadeClassifier will read the file and load the detector

detector = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
allfaces = detector.detectMultiScale(p,1.1)
```

```python
allfaces.shape
```

```python
x,y,w,h = allfaces[0]
```

```python
#cv2.rectangle - parameters are - image source, x and y coords, rectangle shape, 
#color of the rectangle, thicknesss

p = cv2.rectangle(p,(x,y),(x+w,y+h),(255,0,0),2)  # it will select only one face
```

```python
plt.imshow(p)
```

# Detecting all faces

```python
#plotting rectangle for all the images
for face in allfaces:
    x,y,w,h = face
    p = cv2.rectangle(p,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
plt.imshow(p)
```
