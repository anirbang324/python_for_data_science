# openCV 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading an image using opencv
```

```python
img = cv2.imread("./th.jpg")
```

```python
#showing the image
plt.imshow(img)
```

```python
#opening the image in a window for 5 seconds

cv2.imshow("random image",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
#colorspace - normally openCV support colors in BGR format, which alters the colors of the image. But there are several 
# methods in openCV that we can use to change color

new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(new_img)
```

```python
#changing the color to grayscale

img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
# plt.imshow(img_gray)
#to convert the image color properly we should use cmap

plt.imshow(img_gray,cmap="gray")
```

```python
#resizing
#resize takes two parameters - image source, height & width

img_new = cv2.resize(img, (400,400))
```

```python
img_new.shape
```

```python
plt.imshow(img_new)
```

```python
#Blurring an image

img_blur = cv2.blur(img, (12,12))   # 12 is the pixel size
```

```python
plt.imshow(img_blur)
```

```python
#Rotating image
#to rotate an image, we need to go to its centre, so rows/2 and columns/2 will take us to its centre
rows, cols = img.shape[:-1]
im2 = cv2.getRotationMatrix2D((rows/2,cols/2), 30,1)  #last two parameters are degree of rotation and image size percentage
#im2 = cv2.getRotationMatrix2D((rows/2,cols/2), -30,1)  #rotating 30 degree anticlockwise
```

```python
i = cv2.warpAffine(img,im2,(cols,rows))
```

```python
plt.imshow(i)
```

```python
#image translation - its a process by which we can translate an image by few pixel up or few pixel down

M = np.float64([[1,0,50],[0,1,20]])

#M = np.float64([[1,0.1,50],[0,1,-200]])
```

```python
M
```

```python
i = cv2.warpAffine(img,M,(rows,cols))
```

```python
plt.imshow(i)
```
