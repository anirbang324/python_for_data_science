# handling missing data
# dropna() - we can drop those data where we have missing values.
# fillna() - we can approximate the missing values.

```python
iris = pd.read_csv('./iris.csv')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
iris
```

```python
import numpy as np
nan_indx = np.random.randint(0,150,20) # picking up some random values to make them empty
iris['sepal_length'][nan_indx] = np.nan
```

```python
iris['sepal_length'][:50]
```

```python
import numpy as np
nan_indx = np.random.randint(0,150,20) 
iris['petal_length'][nan_indx] = np.nan
```

```python
iris['petal_length'][:50]
```

```python
iris.isna() #checking if the rows are containing NaN values
```

```python
iris.isna().sum()
```

```python
# iris.dropna() # all the rows containing NaN values will be dropped
# dropna() method is not recommended as it will delete the data.
```

```python
# to get the approximate value we should find the mean of sepal_length
iris['sepal_length'] = iris['sepal_length'].fillna(value=round(iris['sepal_length'].mean(),1))
```

```python
iris['sepal_length']
```
